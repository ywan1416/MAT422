# 3.3.2. Convexity and global minimizers

3.3.2.1 Convex Sets and Functions

A convex function has a ‚Äúbowl shape‚Äù meaning that the function curves upwards. For convex functions, any local minimizer is also a global minimizer.

3.3.2.2 Global Minimizers of Convex Functions

For a convex function, the local minimizer is always the global minimizer, and at this point, the gradient or first derivative equals zero

```python
# Define the convex function
def f(x):
    return x**2+x
x_values = np.linspace(-5, 5, 100)
y_values = f(x_values)

# Plot the function
plt.figure(figsize=(8,6))
plt.plot(x_values, y_values, label="f(x) = x^2+x")
plt.scatter(-1/2, f(-1/2), color='red', label="Global Minimizer (x=-1/2, f(x)=-1/4)")
plt.title("Convex Function: f(x) = x^2+x")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

# 3.3.3. Gradient descent

A strongly convex function curves upwards and has a steeper curvature than a regular convex function. This property ensures that the function has only one minimum, which is both a local and global minimizer.

3.3.3.1 Steepest Descent

```python
# Define the strongly convex function
def f(x):
    return x**2 + 2*x + 1
x_values = np.linspace(-4, 2, 100)
y_values = f(x_values)

# Plot the function
plt.figure(figsize=(8,6))
plt.plot(x_values, y_values, label="f(x) = x^2 + 2x + 2")
plt.scatter(-1, f(-1), color='red', label="Global Minimizer (x=-1, f(x)=1)")
plt.title("Strongly Convex Function: f(x) = x^2 + 2x + 2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```
